## Drop Missing Data

```python
# Drop rows containing NaN
df.dropna()
```

```python
#Drop columns containing NaN
df.dropna(axis=1)
```

## Fill the NaNs

```python
# Fill with a fixed value
df.fillna(1000)
```

```python
# Fill Forward
df.fillna(method='ffill')
```

```python
# Fill Backward
df.fillna(method='bfill')
```

```python
# Interpolate values
df.interpolate(method='linear')
```

```python
#Fill with mean
df.fillna(value=df.mean())
```
