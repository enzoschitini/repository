```python
import psutil
import time
import threading

class MonitorCPU(threading.Thread):
    def __init__(self, interval=1):
        super().__init__()
        self.interval = interval
        self.keep_running = True

    def run(self):
        while self.keep_running:
            cpu_percent = psutil.cpu_percent()
            print(f"L'utilizzo della CPU è: {cpu_percent}%")
            time.sleep(self.interval)

    def stop(self):
        self.keep_running = False

# Crea e avvia il thread di monitoraggio
monitor = MonitorCPU()
monitor.start()

# Esegui il tuo programma qui...

%run prestazioni.py

# Quando hai finito, ferma il monitoraggio
monitor.stop()
monitor.join()
```

class Pacchetti(object):
    def __init__(self, frame, colonne:list, anno_di_ricerca:None) -> None:
        self.frame = frame
        self.colonne = colonne
        self.anno_di_ricerca = anno_di_ricerca

    def smistamento(self):
        data_base = self.frame[self.colonne]
        return data_base

    def gli_anni(self):
        anni = self.frame[['date']]
        anni = anni.values.tolist()
        anni_lista = []
        for x in anni:
            anno = x[0]
            anno = anno.split(sep='-')[0]
            anni_lista.append(anno)
        anni_lista = list(set(anni_lista))
        anni_lista.sort()
        return anni_lista

    def affettare(self):
        raccolta = self.smistamento()
        raccolta = raccolta.values.tolist()
        pacchetto = []

            periodo = x[0]
            periodo = periodo.split(sep='-')[0]
            if periodo == self.anno_di_ricerca:
